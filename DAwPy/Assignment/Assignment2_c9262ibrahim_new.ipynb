** Convert datatype of InvoiceDate to datetime and assign it to new column name InvoiceDate2**

```python
ecom["InvoiceDate2"]= pd.to_datetime(ecom["InvoiceDate"])
ecom["InvoiceDate2"]
```

** Take only date values (year-month-day) from InvoiceDate2 and assign it to a new column named date **

```python
ecom["date"]=pd.to_datetime(ecom['InvoiceDate2']).dt.normalize()# bu şekilde sadece tarih alınıyor.
ecom.date
```

```python
ecom.head(5)
```

** drop InvoiceDate2 and InvoiceDate columns permanently **

```python
ecom.drop(labels=["InvoiceDate2","InvoiceDate"],axis=1,inplace=True) # axis 0 satırlar axis 1 sütunlar üzerinden işlem yapıyor.

```

** take only Total and date columns with loc and craete a new df named ecom2 **

```python
ecom2= ecom[['total','date']] # loc olmadan çözümü çözümü...
ecom2
```

```python
ecom2=ecom.loc[:,["total","date"]]
ecom2
```

** take only InvoiceNo and CustomerID columns with iloc and craete a new df named ecom3 **

```python
ecom3 = ecom.iloc[:,[0,5,7]]
ecom3
```

** Combine ecom2 and ecom3 with concat **

```python
pd.concat([ecom2,ecom3],axis=1) # concat metodu aynı boyutta(satır sayıları) aynı olan iki df leri birleştirme için kullanılır. 
```

** **kalın metin** Combine ecom2 and ecom3 with merge **

```python
Combinemerge = pd.merge(ecom2, ecom3, on='total')
Combinemerge
```

** Combine ecom2 and ecom3 with join **

```python
ecom2.join(ecom3, lsuffix='_left', rsuffix='_right') # totall left ve total right  sollu sağlı eklendi. 
```
