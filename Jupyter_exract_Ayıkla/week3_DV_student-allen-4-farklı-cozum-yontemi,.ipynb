Bar plots

````python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
````

# 2. Make a bar plot

````python
df.plot.bar()
````

other solutions:
```
    -# pandas method - selection
    -# functional method
    -# Object Oriented method
```

````python
df.reset_index()
````

````python
df.reset_index().plot(x='index', y=['income', 'expense'], kind='bar')
````

data : indexable object, optional
    An object with labelled data. If given, provide the label names to
    plot in *x* and *y*.

    .. note::
        Technically there's a slight ambiguity in calls where the
        second label is a valid *fmt*. ``plot('n', 'o', data=obj)``
        could be ``plt(x, y)`` or ``plt(y, fmt)``. In such cases,
        the former interpretation is chosen, but a warning is issued.
        You may suppress the warning by adding an empty format string
        ``plot('n', 'o', '', data=obj)``.

````python
# functional method
fig = plt.figure()
plt.subplot()

p = np.arange(len(df))
width = 0.30

plt.bar(p - width/2, df.income, width, label="income")
plt.bar(p + width/2, df.expense,width, label="expense")

plt.xticks(p, df.index.values, rotation='vertical')

plt.legend()
plt.show()
````

````python
# Object Oriented
fig, ax = plt.subplots()

p = np.arange(len(df))
width = 0.30

ax.bar(p - width/2, df.income, width, label="income")
ax.bar(p + width/2, df.expense,width, label="expense")

ax.set_xticks(p, df.index.values, rotation='vertical')
# ax.set_xticklabels(df.index.values)

ax.legend()
plt.show()
````

# 3. Stack the bars

````python
df.plot.bar(stacked=True)
````

other solutions:

````python
df.reset_index().plot(x='index', y=['income', 'expense'], kind='bar', stacked=True)
````

````python
# Object Oriented
fig, ax = plt.subplots()

ax.bar(x=df.index.values, height="income",   width = 0.50, data=df, label="income ")
ax.bar(x=df.index.values, height=df.expense, width = 0.50, label="expense", bottom=df.income)

p = np.arange(len(df))
ax.set_xticks(p, df.index.values, rotation='vertical')

ax.legend()
plt.show()
````

# 11. Make the same plot using seaborn

````python
tips.boxplot(column='total_bill', by='size')
````

````python
# functional method
for n, grp in tips.groupby('size'):
    plt.boxplot(x='total_bill', data=grp, positions=[n])
    
plt.xlabel('size')
plt.ylabel('total_bill')
````

````python
# Object Oriented method
fig, ax = plt.subplots()

for n, grp in tips.groupby('size'):
    ax.boxplot(x='total_bill', data=grp, positions=[n])
    
ax.set_xlabel('size')
ax.set_ylabel('total_bill')
````

````python
# pandas violin plot yok
````

````python
# functional method
for n, grp in tips.groupby('size'):
    plt.violinplot('total_bill', data=grp, positions=[n])
    
plt.xlabel('size')
plt.ylabel('total_bill')
````
